```python
#Â Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
file_path = "/crimes.csv"
crimes = pd.read_csv(file_path, parse_dates=["Date Rptd", "DATE OCC"], dtype={"TIME OCC": str})
crimes.head()
```

TODO1: 1. Finding the frequencies of crimes by the hour of occurrence

```python
crimes["HOUR OCC"] = crimes["TIME OCC"].str[:2].astype(int)
sns.countplot(x="HOUR OCC", data=crimes)
plt.show()
```

```python
peak_crime_hour = int(crimes['TIME OCC'].str[:2].astype(int).mode()[0])
peak_crime_hour
```

TODO2: Identifying the area with the most night crime

```python
night_start = 22
night_end = 3
night_crimes = crimes[(crimes['HOUR OCC'] >= night_start) | (crimes['HOUR OCC'] <= night_end)]
night_crime_counts = night_crimes.groupby("AREA NAME")["DR_NO"].count().sort_values(ascending=False)
peak_night_crime_location = night_crime_counts.index[0]
peak_night_crime_location
```

TODO3: Number of crimes by age group

```python
# Create age bins and labels
age_bins = [0, 17, 25, 34, 44, 54, 64, np.inf]
age_labels = ["0-17", "18-25", "26-34", "35-44", "45-54", "55-64", "65+"]

# Categorize victims by age group
crimes['Age Bracket'] = pd.cut(crimes['Vict Age'], bins=age_bins, labels=age_labels, right=True)

# Count crimes per age group
victim_ages = crimes.groupby('Age Bracket')['DR_NO'].count()

victim_ages
```
